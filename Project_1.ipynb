#### The main goal of this task is to find pairwise correlations and independences for the given probabilities.  

To determine whether x_i and x_j are independent , we check if

\begin{align}
p(x_i, x_j) = p(x_i)p(x_j)
\end{align} 

, where
The joint probability can be calculated by:</br>

\begin{align}
p(x_i, x_j) = p(x_i|x_j)p(x_j)
\end{align}
